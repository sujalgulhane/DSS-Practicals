```python
import numpy as np
from matplotlib import pyplot as plt 
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = 2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Scatter Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
